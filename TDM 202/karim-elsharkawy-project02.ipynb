```python
import requests
from bs4 import BeautifulSoup

# Define url as the url page we want to parse
url = 'https://books.toscrape.com/'
response = requests.get(url)

# response.content gives the content inside the <response> tag
soup = BeautifulSoup(response.content,'html.parser')

# .prettify() formats the code so it's easier to read
print(soup.prettify())
```

I imported requests and bs4. I first use the requests library to get the url from the website (Books To Scrape), then use bs4 to scrape that content, and finally use .prettify in bs4 to format the data and make it easier to read
